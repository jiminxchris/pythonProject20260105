```python
def is_prime(num_list):
    result = []
    for num in num_list:
        isprime = True
        for i in range(2, num):
            if num % i == 0:
                isprime = False
                break
        result.append(isprime)
    return result
```

```python
# 등수 구하기 - 평균점수(영어점수, 수학점수)
def rank(score):
    avg = [sum(s)/len(s) for s in score]
    avg_sort = sorted(avg, reverse=True)
    result = [avg_sort.index(i)+1 for i in avg] # enumerate로 속도 향상!!!!
    return result
```

```python
score1 = [[80, 70], [90, 50, 30], [40, 70, 100, 100], [50, 80]]
rank(score1) # [1, 2, 4, 3]
```

```python
score2 = [[80, 70], [70, 80], [30, 50], [90, 100], [100, 90], [100, 100], [10, 30]]
rank(score2) # [4, 4, 6, 2, 2, 1, 7]
```

```python
class Person:
    def __init__(self, name, age ):
        self.name = name
        self.age = age

    def walk(self):
        print(f'{self.name}이 걸어가요')
```

```python
person1 = Person("박지민", 31)
```

```python
person2 = Person("전정국", 29)
```

```python
# 상속
class NewPerson(Person):
    def fly(self): # 기능확장
        print(f'{self.name}이 하늘을 날아요')
    def walk(self): # 메서드 오버라이딩
        print(f'{self.name}이 아주 빠르게 걸어가요')
```

```python
person3 = NewPerson("민윤기", 33)
```

```python
person3.walk()
```

```python
class NewNewPerson(Person):
    populataion = 0

    def __init__(self, name, age, hobby ):
        super().__init__(name, age)
        self.hobby = hobby
        NewNewPerson.populataion += 1

    def fly(self): # 기능확장
        print(f'{self.name}이 하늘을 날아요')

    def walk(self): # 메서드 오버라이딩
        print(f'{self.name}이 아주 빠르게 걸어가요')

    def introduce(self):
        print(f'{self.name}의 취미는 {self.hobby}입니다. ')
```

```python
person1 = NewNewPerson("전정국", 31, "춤추기")
person2 = NewNewPerson("박지민", 29, "운동하기")
person3 = NewNewPerson("김태형", 33, "노래부르기")
```

```python
person3.populataion
```
