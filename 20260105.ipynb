```python
# 업앤다운
import random
answer = random.randint(1, 100)
print(answer)

while True:
    # 여기에 코드를 구현해주세요.
    # 사용자로 입력(숫자)
    # 조건문(정답, 너무큰지, 너무작은지)
    guess = int(input("1~100사이의 정수를 입력하세요: "))
    if answer == guess:
        print("정답입니다.")
        break
    elif answer < guess:
        print("너무 커요")
    else:
        print('너무 작아요')
print('게임종료')
```

```python
n = 30
star_count = 0

for i in range(1, 11):
    # print('*' * i)
    star_remain = n - star_count
    if star_remain > i:
        print( "*" * i)
        star_count += i
    else:
        print("*" * star_remain)
        break
```

```python
marks = [90, 25, 67, 45, 80]

# 아래와 같이 출력되도록 코드작성 60
1번 학생은 합격입니다.
2번 학생은 불합격입니다.
3번 학생은 합격입니다.
4번 학생은 불합격입니다.
5번 학생은 합격입니다.
```

```python
marks = [90, 25, 67, 45, 80]
num = 0
for i in marks:
    num += 1
    if i >= 60:
        print(f"{num}번 학생은 합격입니다.")
    else:
        print(f"{num}번 학생은 불합격입니다.")
```
