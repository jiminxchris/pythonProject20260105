```python
# 도전미션 
리스트 numbers 각 원소는 숫자 0부터 9까지로 이루어져 있습니다. 이때, numbers에서 연속적으로 나타나는 숫자는 하나만 남기고 전부 제거하려고 합니다. 단, 제거된 후 남은 수들은 numbers의 원소의 순서를 유지해야 합니다. 

예를 들면,
numbers = [1, 1, 3, 3, 0, 1, 1] ->  [1, 3, 0, 1] 
numbers = [4, 4, 4, 3, 3] ->  [4, 3]
코드를 작성해 주세요
```

```python
numbers = [4, 4, 4, 3, 3]

answer = []
compare_num = numbers[0]
answer.append(compare_num)

for i in numbers:
    if compare_num != i:
        compare_num = i
        answer.append(compare_num)

answer
```

```python
# 전화번호 검색 프로그램

addr = {
    '박지민': '010-1111-1234', '전정국': '010-2222-1234',
    '민윤기': '010-3333-1234', '정호석': '010-4444-1234',
    '김태형': '010-5555-1234', '김석진': '010-6666-1234'
}

while True:
    menu = int(input('메뉴선택(1:추가, 2:검색, 0:종료)? '))
    if menu == 1: # 추가기능
        # print('추가하는 기능')
        addr_key = input('이름입력: ')
        addr_value = input('전화번호 입력: ')
        addr[addr_key] = addr_value
    elif menu == 2:# 검색기능
        print("친구들 목록: ", list(addr.keys()))
        addr_name = input('이름 입력:')
        if addr_name in addr:
            print(addr_name, addr[addr_name])
        else:
            print('그런친구는 없습니다. ')
    elif menu == 0:# 종료
        break
    else: # 잘못된메뉴선택
        print("제대로 된 번호를 선택하세요!!")
```

```python
def max_num(a, b, c):
    max_val = a
    if b > max_val:
        max_val = b
    if c > max_val:
        max_val = c
    return max_val
```

```python
max_num(-3, 0, 1, 2)
```

```python
def min_num(array):
    min_val = array[0]
    for i in array:
        if min_val > i :
            min_val = i
    return min_val
```

```python
min_num([0, -1, 3, 87, -100])
```

```python
def isLeafYear(year):
    # year % 4 == 0
    # year % 100 != 0
    # year % 400 == 0
    return (year % 4 == 0) and (year % 100 != 0) or (year % 400 == 0)
```

```python
def add_many(*args):
    # print(args)
    # 전체합을 반환하는 코드
    result = 0
    for i in args:
        result += i
    return result
```

```python
def add_mul(choice, *args):
    print(choice, args)
    if choice == 'add':
        result = 0
        for i in args:
            result += i
        return result
    elif choice == 'mul':
        result = 1
        for i in args:
            result *= i
        return result
    else:
        print('연산자 종류 확인 요망')
        return None
    
```

```python
add_mul('aaa', 1)
```

```python
add_mul('add', 1, 2, 3)
```

```python
add_mul('mul', 1, 2, 3, 4, 5)
```
